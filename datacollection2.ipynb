```python
import cv2
import mediapipe as mp 
import numpy as np
import pandas as pd
import os
from pathlib import Path
```

```python
hand_detector = mp.solutions.hands.Hands()
drawing_utils = mp.solutions.drawing_utils

```

```python

# csv_file_name = "lstm_data.csv"
features_file_name = "pose_features.npy"
labels_file_name = "pose_labels.npy"
label = 7
features = []
video = []
labels = []
```

```python

cap = cv2.VideoCapture(0)

while True :

    _,frame = cap.read()
    frame = cv2.flip(frame , 1)
    rgb_frame = cv2.cvtColor(frame , cv2.COLOR_BGR2RGB)
    output = hand_detector.process(rgb_frame)
    hands = output.multi_hand_landmarks
    fheight , fwidth , _ = frame.shape
    
    if hands :
        hand = hands[0]
        drawing_utils.draw_landmarks(frame , hand , mp.solutions.hands.HAND_CONNECTIONS )
        videofeatures = []
        landmarks = hand.landmark
        
        for id , landmark in enumerate(landmarks):
            x = int(landmark.x*fwidth)
            y = int(landmark.y*fheight)
            cv2.circle(frame , (x,y) , 5 , (255,0,0) , 3)
            videofeatures.append(landmark.x)
            videofeatures.append(landmark.y)
        
        video.append(videofeatures)
        if(len(video) == 15):
            features.append(video)
            labels.append(label)
            video = []
            print(len(features))
            

        
    

        


    cv2.imshow("myhands",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        if(Path(features_file_name).is_file() and Path(labels_file_name).is_file()):
            feature_data = np.load(features_file_name , allow_pickle = True)
            label_data = np.load(labels_file_name , allow_pickle = True)
            feature_data = np.concatenate([feature_data , features])
            label_data = np.concatenate([label_data , labels])
            np.save(features_file_name , feature_data)
            np.save(labels_file_name , label_data)
            
        else :
            np.save(features_file_name , np.array(features))
            np.save(labels_file_name , np.array(labels))
        break
cap.release()
cv2.destroyAllWindows()
```
